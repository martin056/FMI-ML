````python
# Yeah, we have some strings in the data. It's time for encoding.
# I will encode by hand the `playoffs` value because it's "In Playoffs/Not In Playoffs" -> "1/0"

playoffs_fix_data = original_data.copy()

playoffs_fix_data['in_playoffs'] = (playoffs_fix_data.playoffs == 'playoffs').astype(int)

# I will also remove `play` because we don't really need it
playoffs_fix_data = playoffs_fix_data.drop(['playoffs', 'play'], axis=1)

playoffs_fix_data
````

**Next steps:**
In the NBA the quarter time is 12 minutes. As we can see from the dataframe we have to multiply the `period` with `(12:00 - time)`. Let's make a new column `game_time`! The game time deffinetely affects the player's condition and shooting form.

> Just realised I can use Markdown instead of Python comments .........

````python
game_time_data = playoffs_fix_data.copy()

from datetime import datetime

def get_game_time(period, time):
    time_in_table = datetime.strptime(time, '%M:%S').time()
    
    time_in_table_to_seconds = time_in_table.minute * 60 + time_in_table.second
    period_to_seconds = int(period) * 12 * 60 # 1 period = 12 minutes
    
    value = period_to_seconds - time_in_table_to_seconds
    
    return value


game_time_data['game_time'] = game_time_data.apply(lambda r: get_game_time(r['period'], r['time']), axis=1)
game_time_data = game_time_data.drop(['period', 'time'], axis=1)

game_time_data
````

Good, now we have the exact time when the free throws are commited.

What we forgot? The `game_id` is redundant. We will remove it.

````python
data = game_time_data.copy()

data = data.drop(['game_id'], axis=1)
data
````

**What's next?** Well, as we can see there are still some strings in the dataframe. Let's get rid off the `end_result` now! What actually might be important is if we **won or lost** the game. I'll make columns for this and will remove `end_result`.

Plan:

* Make function that calculates the end result - possitive/negative integer
* Change the type of `end_result` column
* Depending on it create `won_game` and `lost_game` columns
* Don't drop `end_result` - the score may addect the prediction

````python
def get_game_result(end_result):
    end_result_pair = end_result.split(" - ")
    guest_result, host_result = tuple(end_result_pair)  # in NBA the result is opposite, e.g. LAL @ PHX
    
    return int(host_result) - int(guest_result)  # will be negative if we lost the game

fix_end_result_data = data.copy()
fix_end_result_data['end_result'] = fix_end_result_data.apply(lambda r: get_game_result(r['end_result']), axis=1)

fix_end_result_data['game_lost'] = (fix_end_result_data.end_result < 0).astype(int)
fix_end_result_data['game_won'] = (fix_end_result_data.end_result > 0).astype(int)

fix_end_result_data
````

What's left? Ok, there is a `score`. Can he affect the occurancy of a shot? I guess **yes** - if the game is tied the player is under a higher pressure. Let's use the function from above and fix this column.

````python
fix_score_data = fix_end_result_data.copy()
fix_score_data['current_score'] = fix_score_data.apply(lambda r: get_game_result(r['score']), axis=1)
fix_score_data = fix_score_data.drop(['score'], axis=1)

fix_score_data
````

## But ....

What do we really predict? In the beginning I told myself let's predict if the baller is going to score - classification. Later, I decided to get what is the chance for the baller to score (regression model). Yeah, that's great but our data at the moment is not very suitable - put `y` is `shot_made`. What's the problem?

The problem is that the baller can shoot **1, 2 or 3** free throws depending on the file type. The `shot_made` currently tells us "the baller has scored a point". In other words, if player A scored 1 from 1 foul and player B has scored 1 of 3 fouls - the both have `score_made = 1` which will break the regression model.

## Go back

What can we do? Let's go back to the original data. There is a column `play` where we can find a strinf like this:
`Andrew Bynum makes free throw 1 of 2`. We need the "makes/misss" and the numbers from this string.

![](http://media.pyweek.org/dl/7/RegExExpress/train_anim2.gif)

````python
import re

def split_play(play):
    action = re.search('(misses|makes)', play)
    action = action.group(1)
    
    numbers = re.findall('\d+', play)
    x, y = tuple(numbers)
    
    x = int(x)
    y = int(y)
    
    if action == 'makes':
        return x / y
    
    if action == 'misses':
        return (y - x) / x
    
    
# drop_season_data['shot_made_%'] = original_data.apply(lambda r: split_play(r['play']), axis=1)
````

````python
import re

def split_play(play):
    action = re.search('(misses|makes)', play)
    action = action.group(1)
    
    numbers = re.findall('\d+', play)
    
    if len(numbers) == 0:
        return action == "makes" and 1.00 or 0.00
    
    if len(numbers) == 1:
        return 0.00

    x, y = tuple(numbers)
    
    x = int(x)
    y = int(y)
    
    if action == 'makes':
        return round((x / y), 2)
    
    if action == 'misses':
        return round(((y - x) / y), 2)
    
fix_shot_made_data = drop_season_data.copy()

fix_shot_made_data['shot_made_%'] = original_data.apply(lambda r: split_play(r['play']), axis=1)
fix_shot_made_data = fix_shot_made_data.drop(['shot_made'], axis=1)
fix_shot_made_data
````

*I will use the code from above*

````python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder


categorical_columns = ['game', 'player']
encoders = {col: LabelEncoder().fit(fix_shot_made_data[col]) for col in categorical_columns}

def encode_categorical(data, columns, encoders):
    data = data.fillna("")
    return pd.DataFrame({col: encoders[col].transform(data[col]) for col in columns},
                        index = data.index)

encoded_data = encode_categorical(fix_shot_made_data, categorical_columns, encoders)

one_hot = OneHotEncoder().fit(encoded_data)
print(one_hot.transform(encoded_data).shape)
````

````python
final_data = fix_shot_made_data.copy()

y = final_data['shot_made_%']
X = final_data.drop(['shot_made_%'], axis=1)
````

Our data can't use a linear regression - we have **only 5 types of values** in the `y` - 0, 0.33, 0.5, 0.67, 1.0 ...

````python
def convert_shot_made(str_):
    options = {
        0.0: '0%',
        0.33: '33%',
        0.5: '50%',
        0.67: '67%',
        1.0: '100%'
    }
    
    return options[str_]

y_cat = y.apply(lambda val: convert_shot_made(val))
y_cat
````

````python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.3, random_state=33, stratify=y_cat)
one_hot_x_train = one_hot.transform(encode_categorical(X_train[categorical_columns], categorical_columns, encoders))
one_hot_x_test = one_hot.transform(encode_categorical(X_test[categorical_columns], categorical_columns, encoders))
````

````python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression().fit(one_hot_x_train, y_train)

print("Logistic Regression")
print("Train set: ", logistic_regression.score(one_hot_x_train, y_train))
print("Test set: ", logistic_regression.score(one_hot_x_test, y_test))
````

````python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(
    random_state=33,
    max_depth=5,
    n_estimators=100
)

random_forest.fit(one_hot_x_train, y_train)

print("Random Forest")
print("Train set: ", random_forest.score(one_hot_x_train, y_train))
print("Test set: ", random_forest.score(one_hot_x_test, y_test))
````

### Let's play with the Random Forest now

````python
from sklearn.ensemble import RandomForestClassifier
import random

n_estimators = [10, 100, 1000]
max_depth = [5, 6, 7]
min_samples_split = [2, 3, 4, 5]


for _ in range(10):
    est = random.choice(n_estimators)
    depth = random.choice(max_depth)
    splt = random.choice(min_samples_split)

    random_forest = RandomForestClassifier(
        random_state=33,
        max_depth=depth,
        n_estimators=est,
        min_samples_split=splt,
        n_jobs=3
    )

    random_forest.fit(one_hot_x_train, y_train)

    print("Random Forest with estimators: {}, max_depth: {}, min_samples_split: {}".format(depth, est, splt))
    print("Train set: ", random_forest.score(one_hot_x_train, y_train))
    print("Test set: ", random_forest.score(one_hot_x_test, y_test))
    print('\n')
````

## Finally

I really want to check what is the score if we remove the two outlier classes (`0.67` & `0.33`).

Our prediction still is going to be valid - as we can see from the plot above, very few 3-attempt fouls happens.

````python
data_with_no_outliers = fix_shot_made_data.copy()

def remove_outliers(shot_made):
    if shot_made == 0.67 or shot_made == 0.33:
        return 0.5
    
    return shot_made

data_with_no_outliers['shot_made_%'] = data_with_no_outliers.apply(lambda r: remove_outliers(r['shot_made_%']), axis=1)

y = data_with_no_outliers['shot_made_%']
X = data_with_no_outliers.drop(['shot_made_%'], axis=1)

y = y.apply(lambda val: convert_shot_made(val))

y.value_counts().plot(kind='bar');
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=33, stratify=y)

final_x_train = one_hot.transform(encode_categorical(X_train[categorical_columns], categorical_columns, encoders))
final_x_test = one_hot.transform(encode_categorical(X_test[categorical_columns], categorical_columns, encoders))


final_logistic_regression = LogisticRegression(C=100000).fit(final_x_train, y_train)

print("Logistic Regression")
print("Train set: ", final_logistic_regression.score(final_x_train, y_train))
print("Test set: ", final_logistic_regression.score(final_x_test, y_test))
````

# Conclusion

Well, that was my first data exploration. Let me summarize what I've learned:

* Data exploration is a process full of disappointment, frustration and anger but you feel relaxed in the end.
* You can combine Markdown and Python code in your notebooks :D
* I haven't used Pandas, numpy and scikitlearn before (and they are awesome)

* You have to be very careful with what you are trying to predict - the data may look suitable for a regression but you had better use a classification instead!
